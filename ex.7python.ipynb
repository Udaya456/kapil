```python
# Define the range and step size
x =np.arange(-100, 100, 0.01)

# Define the function y = x^2 + x - 11
y = x**2 + x - 11

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = x^2 + x - 11")

# Set limits for x and y for zooming in to find the root
plt.xlim(-5, 5)  # Narrow range around where the roots likely are
plt.ylim(-1, 1)  # Focus around the x-axis

# Add title and labels
plt.title("Zooming in to find the roots of y = x^2 + x - 11")
plt.xlabel("x")
plt.ylabel("y")

# Add gridlines
plt.grid(True, which='both', linestyle='--', linewidth=0.7)

# Add horizontal and vertical axis lines at y=0 and x=0
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Show the plot
plt.legend()
plt.show()


```
