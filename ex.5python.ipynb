```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(-100,100,0.01)
y=x**2 +x -11

# create the plot
plt.figure(figsize=(8,6))
plt.plot(x,y, label="y=x**2+x+11")
plt.title("Graph of the function y= x^2+x-11")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.axhline(y=0, color='black', linewidth=0.5)
plt.legend()

#show the plot
plt.show()

 
```
