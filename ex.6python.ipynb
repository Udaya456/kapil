```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range and step size
x = np.arange(-100, 100, 0.01)

# Define the function y = x^2 + x - 11
y = x**2 + x - 11

# Create the plot with limited display range for x and y
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = x^2 + x - 11")

# Set limits for x and y
plt.xlim(-25, 25)
plt.ylim(-100, 600)

# Add title and labels
plt.title("Graph of the function y = x^2 + x - 11")
plt.xlabel("x")
plt.ylabel("y")

# Add gridlines
plt.grid(True, which='both', linestyle='--', linewidth=0.7)

# Add horizontal and vertical axis lines at y=0 and x=0
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Show the plot
plt.legend()
plt.show()
```
